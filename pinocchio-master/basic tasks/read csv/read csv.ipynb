# Read in CSV data

### Data from SQL
Datetimes from the database should be UTC so we'll assume any datetimes in Python are also UTC.

```sql
-- get rental data
select 
  id, customer_id, created_at, cancelled_at, booked_at, end_datetime
from
  carsharing_rental
limit 5000
```

````python
import pandas as pd
````

````python
data_filename = 'rentaldata_2019-4-19_1221.csv'

# read the csv and parse columns 2, 3, 4, and 5, as time zone naive (UTC) datetimes
df = pd.read_csv(data_filename, 
                 parse_dates=[2,3,4,5],
                 infer_datetime_format=True)
df.head()
````

If you want to use or display datetimes in a different timezone, you have to `localize` and then `convert` them

````python
df['end_datetime_pacific'] = df['end_datetime'].dt.tz_localize('UTC').dt.tz_convert('America/Los_Angeles')
df.head()
````
